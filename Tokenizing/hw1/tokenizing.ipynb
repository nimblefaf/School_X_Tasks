```python
import nltk
import nltk.tokenize

import timeit
```

```python
#загружаем все 122 главы в одну переменную
text = ''
for i in range(1, 123):
    path = 'C:\\Users\\user\\Desktop\\DeepAutomation\\Tokenizing\\hw1\\chapters\\' + str(i).rjust(4, '0') + '.txt'
    with open(path) as f:
        text += ' '.join(f.readlines())
len(text)
```

```python
start_time = timeit.default_timer()

whitespace = nltk.tokenize.WhitespaceTokenizer()
tokens_whitespace = whitespace.tokenize(text)

print(timeit.default_timer() - start_time)
```

```python
tokens_whitespace
```

```python
nltk.tokenize.word_tokenize(text)
```

```python
import spacy
```

```python
#spacy ругается на слишком длинный текст, поэтому прогоню частично
len(text) # = 3 752 789

text = text[:999999]
len(text)
```

```python
start_time = timeit.default_timer()

model = spacy.load('en_core_web_sm')
doc = model(text)
tokens = [token.text for token in doc]
print(timeit.default_timer() - start_time)
```

```python
tokens
```

```python
text = text.replace('\n', '')
```

```python
start_time = timeit.default_timer()

model = spacy.load('en_core_web_sm')
doc = model(text)
tokens = [token.text for token in doc]
print(timeit.default_timer() - start_time)
```
