```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

import pandas as pd
import numpy as np

%matplotlib inline
```

Device_id – уникальный идентификатор девайса отправителя события
Device_type – тип девайса
Datetime – время отправки события
Ev_type – тип события (0 – показ, 1 – клик)
Adv_type – тип рекламы
Region_id – регион отправителя события
Age – возраст отправителя события
Gender – пол отправителя события (1 – женщины, 2 – мужчины)

```python
data = pd.read_csv("python_task_dataset.csv")
data.head(10)
```

```python
data_clicks = data[data.ev_type > 0]
```

```python
data_clicks.head(10)
```

```python
fig = px.scatter(
    data_clicks, 
    y='age', x='gender', 
    color='device_type',
    opacity=.3
)
fig.show()
```

```python
fig = px.scatter(
    data_clicks, 
    y='age', x='device_id', 
    color='device_type',
    opacity=.3
)
fig.show()
```

```python
fig = px.scatter(
    data_clicks, 
    y='age', x='datetime', 
    color='device_type',
    opacity=.3
)
fig.show()
```

```python
data_clicks_males = data_clicks[data_clicks.gender == 2]
```

```python
fig = px.scatter(
    data_clicks_males, 
    y='age', x='datetime', 
    color='device_type',
    opacity=.3
)
fig.show()
```

```python
data_clicks_females = data_clicks[data_clicks.gender == 1]
```

```python
fig = px.scatter(
    data_clicks_females, 
    y='age', x='datetime', 
    color='adv_type',
    opacity=.3
)
fig.show()
```
