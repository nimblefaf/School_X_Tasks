```python
import random

# определяю операции с векторами для более аккуратного кода
class Vector:
    def __init__(self, vector: list):
        self.vect = vector

    def __repr__(self) -> str:
        return str(self.vect)
    
    def __sub__(self: 'Vector', other: 'Vector') -> 'Vector':
        #assuming that vectors have the same size
        result = Vector([])
        for i in range(len(self.vect)):
            result.vect.append(self.vect[i] - other.vect[i])
        return result
    
    def __mul__(self: 'Vector', other: 'Vector|int|float') -> 'Vector':
        result = Vector([])
        if isinstance(other, Vector):
            raise NotImplementedError("Умножениие вектора на вектор не доделано")
            return result # not needed for now
        else:
            for i in range(len(self.vect)):
                result.vect.append(self.vect[i] * other)
            return result
        
    def __matmul__(self: 'Vector', other: 'Vector') -> int|float: # даём оператору @ функцию скалярного умножения
            result = 0
            for i in range(len(self.vect)):
                result += self.vect[i] * other.vect[i]
            return result
        
    def is_zero(self) -> bool:
        for elem in self.vect:
            if elem != 0:
                return False
        return True
    
    def norm(self) -> int|float: # ||f1|| = среднеквадратичная от ее координат, или корень от суммы квадратов ее координат
        res = 0
        for i in range(len(self.vect)):
            res += self.vect[i] ** 2
        return res ** 0.5


    
# класс матриц сделанный для дз по питону
class Matrix:
    def __init__(self, matr: list):
        # first_row_size = len(matr[0]) #matrix should have same dimensions (AxA)
        # for row in range(1, len(matr)):
        #     if len(matr[row]) != first_row_size:
        #         raise ValueError("Wrong matrix dimensions!")     # nuke this shit if matrix dimensions are fucked
        self.matrix = matr
        self.rows = len(self.matrix)
        if(type(self.matrix[0]) == list):
            self.columns = len(self.matrix[0])
        else:
            self.columns = 1

    
    def __get__(self):
        return self.matrix
    
    def generate_matrix(rows: int, columns: int, rnd= False) -> 'Matrix':
        result = Matrix([[0] * columns for i in range(rows)])
        if(rnd):
            for row in range(result.rows):
                for column in range(result.columns):
                    result.matrix[row][column] = random.randint(0, 10)
        return result

    def T(self):
        result = Matrix.generate_matrix(self.columns, self.rows)
        for row in range(result.rows):
            for column in range(result.columns):
                result.matrix[row][column] = self.matrix[column][row]
        return result

    def __add__(self: 'Matrix', other: 'Matrix|int|float'):
        result = Matrix.generate_matrix(self.rows, self.columns)
        if isinstance(other, Matrix):
            for row in range(self.rows):
                for column in range(self.columns):
                    result.matrix[row][column] = self.matrix[row][column] + other.matrix[row][column]
        else:
            for row in range(self.rows):
                for column in range(self.columns):
                    result.matrix[row][column] = self.matrix[row][column] + other
        return result
    
    def __sub__(self: 'Matrix', other: 'Matrix|int|float'):
        result = Matrix.generate_matrix(self.rows, self.columns)
        if isinstance(other, Matrix):
            for row in range(self.rows):
                for column in range(self.columns):
                    result.matrix[row][column] = self.matrix[row][column] - other.matrix[row][column]
        else:
            for row in range(self.rows):
                for column in range(self.columns):
                    result.matrix[row][column] = self.matrix[row][column] - other
        return result
    
    def __truediv__(self: 'Matrix', other: 'int|float'):
        result = Matrix.generate_matrix(self.rows, self.columns)
        if isinstance(other, Matrix):
            for row in range(self.rows):
                for column in range(self.columns):
                    result.matrix[row][column] = self.matrix[row][column] / other.matrix[row][column]
            return result
        else:
            for row in range(self.rows):
                for column in range(self.columns):
                    result.matrix[row][column] = self.matrix[row][column] / other
            return result
    
    def __mul__(self: 'Matrix', other: 'Matrix|int|float'):
        result = Matrix.generate_matrix(self.rows, self.columns)
        if isinstance(other, Matrix):
            for row in range(self.rows):
                for column in range(self.columns):
                    result.matrix[row][column] = self.matrix[row][column] * other.matrix[row][column]
            return result
        else:
            for row in range(self.rows):
                for column in range(self.columns):
                    result.matrix[row][column] = self.matrix[row][column] * other
            return result
        
    def __matmul__(self: 'Matrix', other: 'Matrix'):
        result = Matrix.generate_matrix(self.rows, other.columns)
        for row in range(result.rows):
            for column in range(result.columns):
                temp: int|float = 0
                for i in range(self.columns):
                    temp += self.matrix[row][i] * other.matrix[i][column]
                result.matrix[row][column] = temp
        return result
    
    def __eq__(self, other: list) -> bool:
        return self.matrix == other


    def __repr__(self) -> str:
        return str(self.matrix)
    
    # not working like I hoped it would
    def printable(self) -> str:
        result = ""
        for row in range(self.rows):
                result += str(self.matrix[row])
                result += '\n'
        return result

    def GramSchmidt_4x4(self) -> 'Matrix':
        
        # check if size is 4x4
        if(self.rows != 4):
            raise ValueError("")

        # do the math
        v1 = Vector(self.matrix[0])
        v2 = Vector(self.matrix[1])
        v3 = Vector(self.matrix[2])
        v4 = Vector(self.matrix[3])

        #первоначальная попытка наткнулась на проблему деления на ноль
        # f1 = v1

        # a2 = (v2 @ f1) / (f1 @ f1)
        # f2 = v2 - (f1 * a2)

        # a3 = (v3 @ f1) / (f1 @ f1)
        # b3 = (v3 @ f2) / (f2 @ f2)
        # f3 = v3 - (f1 * a3) - (f2 * b3)
        # print(f3.vect)

        # a4 = (v4 @ f1) / (f1 @ f1)
        # b4 = (v4 @ f2) / (f2 @ f2)
        # c4 = (v4 @ f3) / (f3 @ f3)
        # f4 = v4 - (f1 * a4) - (f2 * b4) - (f3 * c4)
        

        # вычисления базиса циклом 
        # (с проверкой на нулевой вектор при вычислении скаляров)
        Vects = [v1, v2, v3, v4]
        orts = []
        for i in range(4):
            new_v = Vects[i]
            for j in range(len(orts)):
                if (not orts[j].is_zero()):
                    new_v -= orts[j] * ((Vects[i] @ orts[j]) / (orts[j] @ orts[j]))
            orts.append(new_v)


        # отбросим нулевые вектора перед выдачей результата
        result = []
        for vector in orts:
            if(not vector.is_zero()):
                result.append(vector.vect)
        return Matrix(result)
    
    def SingularPresentation(self) -> 'Matrix': # привести вектора в матрице к единичному виду
        result = []
        for v in self.matrix:
            vector = Vector(v)
            result.append((vector * (1 / vector.norm())).vect)
        return Matrix(result)
    
```
